```python
import os
import pandas as pd

# Define the path to the datasets folder
data_dir = "datasets"
all_data = []  # List to store DataFrames

# Iterate through each month folder
for month_folder in os.listdir(data_dir):
    month_path = os.path.join(data_dir, month_folder)
    if os.path.isdir(month_path):  # Check if it's a directory
        for file in os.listdir(month_path):
            if file.endswith(".csv"):  # Check for CSV files
                file_path = os.path.join(month_path, file)
                df = pd.read_csv(file_path)  # Load the CSV file
                df['Month'] = month_folder  # Add a column for the month
                df['File'] = file  # Add a column for the file name
                all_data.append(df)

# Combine all data into a single DataFrame
combined_data = pd.concat(all_data, ignore_index=True)

# Save to a new CSV file if needed
combined_data.to_csv("combined_delhi_data_2022.csv", index=False)

print("Data combined successfully!")
```

```python
# Detect and cap outliers
for col in numerical_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    print(f"{col} - Outlier Bounds: Lower = {lower_bound}, Upper = {upper_bound}")

    # Capping the outliers
    df[col] = np.where(df[col] < lower_bound, lower_bound, 
                       np.where(df[col] > upper_bound, upper_bound, df[col]))
```

```python
# Total power consumption across all sources
df['Total_Consumption'] = df[numerical_cols].sum(axis=1)

# Proportion of Delhi's consumption in total
df['DELHI_Ratio'] = df['DELHI'] / df['Total_Consumption']

# Verify new features
print(df[['Total_Consumption', 'DELHI_Ratio']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define regions
regions = ['DELHI', 'BRPL', 'BYPL', 'NDPL', 'NDMC', 'MES']

# Ensure the 'Month' column contains valid month names
print("Unique values in 'Month':", df['Month'].unique())  # Debugging step

# Map months to seasons
season_map = {
    'January': 'Winter', 'February': 'Winter', 'March': 'Spring',
    'April': 'Spring', 'May': 'Summer', 'June': 'Summer',
    'July': 'Monsoon', 'August': 'Monsoon', 'September': 'Autumn',
    'October': 'Autumn', 'November': 'Winter', 'December': 'Winter'
}

# Apply the season mapping
df['Season'] = df['Month'].map(season_map)

# Check if mapping was successful
print("Unique values in 'Season':", df['Season'].unique())  # Debugging step

# Group by 'Season' and calculate mean for the regions
seasonal_data = df.groupby('Season')[regions].mean()

# Check if data exists for each season
print(seasonal_data)  # Debugging step

# Proceed to plot if data is available
if not seasonal_data.empty:
    # Plot seasonal trends
    seasonal_data.plot(kind='bar', figsize=(12, 6), colormap='coolwarm')
    plt.title("Seasonal Trends in Power Consumption Across Regions")
    plt.ylabel("Average Power Consumption")
    plt.xlabel("Season")
    plt.xticks(rotation=45)
    plt.grid()
    plt.show()
else:
    print("No data available to plot. Please check the dataset.")
```

```python
# Create ratios for analysis
df['BRPL_to_BYPL'] = df['BRPL'] / df['BYPL']
df['NDPL_to_NDMC'] = df['NDPL'] / df['NDMC']

# Boxplots for ratios
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['BRPL_to_BYPL', 'NDPL_to_NDMC']])
plt.title("Distribution of Power Ratios Between Regions")
plt.show()
```

```python
# Calculate Coefficient of Variation (CV) for each region
cv_data = {region: (df[region].std() / df[region].mean()) for region in regions}

# Plot CV values
plt.figure(figsize=(10, 6))
sns.barplot(x=list(cv_data.keys()), y=list(cv_data.values()), palette="coolwarm")
plt.title("Coefficient of Variation for Power Consumption Across Regions")
plt.ylabel("Coefficient of Variation")
plt.xlabel("Region")
plt.show()
```

```python

# Define regions
regions = ['DELHI', 'BRPL', 'BYPL', 'NDPL', 'NDMC', 'MES']

# Check unique values in 'TIMESLOT' column
print("Unique time slots:", df['TIMESLOT'].unique())  # Debugging step

# Categorize TIMESLOT into time of day (morning, afternoon, evening, night)
time_of_day_map = {
    "00:00-06:00": "Night",
    "06:00-12:00": "Morning",
    "12:00-18:00": "Afternoon",
    "18:00-24:00": "Evening"
}

# Extract hours from TIMESLOT and map to time of day
def categorize_time_slot(timeslot):
    if "00:00" in timeslot or "01:00" in timeslot or "02:00" in timeslot or "03:00" in timeslot or "04:00" in timeslot or "05:00" in timeslot:
        return "Night"
    elif "06:00" in timeslot or "07:00" in timeslot or "08:00" in timeslot or "09:00" in timeslot or "10:00" in timeslot or "11:00" in timeslot:
        return "Morning"
    elif "12:00" in timeslot or "13:00" in timeslot or "14:00" in timeslot or "15:00" in timeslot or "16:00" in timeslot or "17:00" in timeslot:
        return "Afternoon"
    elif "18:00" in timeslot or "19:00" in timeslot or "20:00" in timeslot or "21:00" in timeslot or "22:00" in timeslot or "23:00" in timeslot:
        return "Evening"
    else:
        return "Unknown"

df['TimeOfDay'] = df['TIMESLOT'].apply(categorize_time_slot)

# Average power consumption by time of day
time_of_day_data = df.groupby('TimeOfDay')[regions].mean()

# Sort by time of day order
time_of_day_order = ['Night', 'Morning', 'Afternoon', 'Evening']
time_of_day_data = time_of_day_data.loc[time_of_day_order]

# Plot power consumption trends by time of day
time_of_day_data.plot(kind='bar', figsize=(12, 6), colormap='viridis')
plt.title("Average Power Consumption by Time of Day")
plt.ylabel("Average Power Consumption")
plt.xlabel("Time of Day")
plt.xticks(rotation=45)
plt.legend(title="Regions", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid()
plt.show()
```

```python

# Plot distribution for each region
plt.figure(figsize=(15, 10))
for i, region in enumerate(['DELHI', 'BRPL', 'BYPL', 'NDPL', 'NDMC', 'MES'], 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[region], kde=True, bins=30, color='skyblue')
    plt.title(f"Distribution of Power Consumption in {region}")
    plt.xlabel("Power Consumption")
    plt.ylabel("Frequency")
    plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Calculate cumulative sum of power consumption over time (across all regions)
df['Cumulative_DELHI'] = df['DELHI'].cumsum()
df['Cumulative_BRPL'] = df['BRPL'].cumsum()
df['Cumulative_BYPL'] = df['BYPL'].cumsum()
df['Cumulative_NDPL'] = df['NDPL'].cumsum()
df['Cumulative_NDMC'] = df['NDMC'].cumsum()

# Plot cumulative power consumption
plt.figure(figsize=(12, 6))
plt.plot(df['Cumulative_DELHI'], label='DELHI', color='b')
plt.plot(df['Cumulative_BRPL'], label='BRPL', color='g')
plt.plot(df['Cumulative_BYPL'], label='BYPL', color='r')
plt.plot(df['Cumulative_NDPL'], label='NDPL', color='orange')
plt.plot(df['Cumulative_NDMC'], label='NDMC', color='purple')

plt.title("Cumulative Power Consumption Across Regions Over Time")
plt.xlabel("Index")
plt.ylabel("Cumulative Power Consumption")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Apply rolling mean (window size 7) to smooth out the time-series data
df['Rolling_DELHI'] = df['DELHI'].rolling(window=7).mean()
df['Rolling_BRPL'] = df['BRPL'].rolling(window=7).mean()
df['Rolling_BYPL'] = df['BYPL'].rolling(window=7).mean()
df['Rolling_NDPL'] = df['NDPL'].rolling(window=7).mean()
df['Rolling_NDMC'] = df['NDMC'].rolling(window=7).mean()

# Plot rolling mean
plt.figure(figsize=(12, 6))
plt.plot(df['Rolling_DELHI'], label='DELHI', color='b')
plt.plot(df['Rolling_BRPL'], label='BRPL', color='g')
plt.plot(df['Rolling_BYPL'], label='BYPL', color='r')
plt.plot(df['Rolling_NDPL'], label='NDPL', color='orange')
plt.plot(df['Rolling_NDMC'], label='NDMC', color='purple')

plt.title("7-Day Rolling Mean of Power Consumption Across Regions")
plt.xlabel("Index")
plt.ylabel("Rolling Mean of Power Consumption")
plt.legend()
plt.grid(True)
plt.show()
```
